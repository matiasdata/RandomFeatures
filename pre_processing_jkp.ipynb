```python
char_cols = chars_cfm_130
```

```python
dates_list = list(data["eom"][(data["eom"] >= datetime.date(1963, 1, 31))
                               & (data["eom"] <= datetime.date(2022, 12, 31))].unique())
dates_list.sort()
```

```python
# Initialize an empty DataFrame to store results
results = pd.DataFrame()

for date in dates_list:
    print(f"Processing date: {date}")
    
    # Filter the data for the current date
    data_period = data[data["eom"] == date]
    
    # Calculate the threshold for non-missing values (at least 70% of 130)
    valid_threshold = int(130 * 0.7)
    
    # Count non-NaN values in each row for char_cols
    non_missing_counts = data_period[char_cols].notna().sum(axis=1)
    
    # Drop rows where non-missing values in char_cols are less than the threshold
    data_period = data_period[non_missing_counts >= valid_threshold].copy()


    # If there are no valid rows after filtering, skip to the next date
    if data_period.empty:
        print(f"No valid data for {date} after filtering for missing values.")
        continue
    
    # Select necessary columns
    df_z = data_period[["id", "eom", "excntry", "gvkey", "permno", "size_grp", "me", "ret_exc_lead1m"]].copy()
    
    # Create a dictionary to store the transformed characteristic columns
    transformed_char_cols = {}
    
    # Process the characteristics for each column in char_cols
    for column in char_cols:
        data_period[column] = pd.to_numeric(data_period[column], errors='coerce')
        rank = data_period[column].rank(method="average", na_option='keep', ascending=False)
        n_t = rank.count()
        rc = rank / (n_t + 1)
        mu = rc.mean()
        transformed_char_cols[column] = (rc - mu) / (np.abs(rc - mu).sum())
    
    # Convert the dictionary to a DataFrame and fill NaN values with 0
    transformed_df = pd.DataFrame(transformed_char_cols).fillna(0)
    
    # Concatenate the transformed columns with the original df_z
    df_z = pd.concat([df_z, transformed_df], axis=1)
    
    # Concatenate the processed data for this date into the results
    results = pd.concat([results, df_z])
    print(df_z.shape)
```

```python
results["R_e_adj"] = results["ret_exc_lead1m"]
results["date"] = results["eom"]
results = results.drop(columns=["ret_exc_lead1m","eom"])
results
```

```python
# there are missing excess returns, proxied to zero
results["R_e_adj"].fillna(0.0, inplace=True)
```
